```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python

# Okay, so I think that what actually happens...
# Is that I was about to reinvent Stochastic! RK4
# And since I don't necessarily have any good theory for that,
# I'll use Euler Maruyama for the solution instead.

def forward_eulerish(y0, t_eval, kmean, step_limiting=False, damping=0.05):
    # stochastic forward euler. t_range isn't used because 
    # the equation is autonomous.
    
    y = y0
    
    k = kmean
    dt = t_eval[1] - t_eval[0]
    
    end = len(t_eval)
    if step_limiting:
        end = step_limiting
    
    ys = np.zeros((2, end))
    for i in range(end):
        ys[:,i] = y
        
        f = lambda x: np.array([x[1], - k * x[0] ])
        
        y = y + dt * f(y)
        
        # k is randomly shifted
        k += damping * np.random.randn() * np.sqrt(dt) # - k * dt
        if k < 0: # clamp damped values
            k = 0.
    
    return ys

t0, tf = 0., 10.
kmean = 1.
t_eval = np.linspace(t0, tf, 100 * 2 ** 10)
y0 = [0., 1.]
ys = forward_eulerish(y0, t_eval, kmean, damping=0.1)
#plt.plot(t_eval, ys[0,:], '-o')
        
```

```python
for i in range(20):
    ys = forward_eulerish(y0, t_eval, kmean, damping=0.1)
    plt.plot(t_eval, ys[0,:], '-')
    ylim = 1.5
    plt.ylim([-ylim,ylim])
    
plt.plot(t_eval, np.sin(t_eval), 'ko')

plt.savefig('assets/sde_realizations.png', dpi=256)
```

# Annealing experiment.

```python
# Generate test data..

t_eval = np.linspace(0, 10, 2 ** 10)
ys = []
for i in range(20):
    ys += [forward_eulerish(y0, t_eval, kmean, damping=0.1)]
# maybe don't regenerate these every time though.

Ys = ys
```

```python
from scipy.optimize import dual_annealing

def curve_diffs(params, T, Ys):
    kmean, damping = params[0], params[1] # damping gives stddev
    
    nsamples = 2 ** 6
    datapts = []
    
    # Based on the parameter guess, run a bunch of samples.
    # (I don't know how else to fit the variance of k,
    # since I can't infer the variance from the solutions to the ODE
    # necessarily. )
    for i in range(nsamples):
        datapt = forward_eulerish(y0, T, kmean, damping=damping)
        datapts.append(datapt)
    
    total_err = 0.
    for datapt in datapts:
        err_data = 0.
        for Y in Ys:
            err_data += np.sqrt( np.sum((datapt - Y) ** 2) )
        total_err += err_data #pls don't overflow..
    
    return total_err
```

```python
%%time 

guess = dual_annealing(curve_diffs, list(zip([0., 0.0], [2., 1.])),
                       args=(t_eval, Ys), maxiter=1)
print(guess)
```

```python
%%time 


guess = dual_annealing(curve_diffs, list(zip([0., 0.0], [2., 1.])),
                       args=(t_eval, Ys), maxiter=60)
print(guess)
```

```python
# Paranoid k checking
# The mean of the clamped R.V.

ki = 1.
nframes = 2 ** 6
t0, tf = 0., 5.

dt = (tf - t0) / (nframes - 1)

nsamples = 2 ** 10
ks = []
for j in range(nsamples):
    damping = 0.05
    
    k = ki
    for i in range(nframes):
        # constant in front because I **don't** want the noise to dominate the timescale (ahh)
        k += damping * np.random.randn() * np.sqrt(dt) 
        
        if k < 0: # clamp damped values
            k = 0.
    ks += [k]

# for the unclamped variable k, the variance is at np.sqrt(tf) (very high!) 
# The clamped variable has lower variance.
# But I should also 
print(np.mean(ks), np.median(ks), np.std(ks)) 

# Also plot the histogram of ks.
plt.hist(ks)


#print(ks)
```
