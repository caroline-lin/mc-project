```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
'''
April 2nd

Adapt van_der_pol code for displaying phase plane of oscillator.
'''
```

```python
mu = 0.5

# Shared mu, k = 1
f_harmonic = lambda t, x: [ x[1], mu * x[1] - x[0] ]

f_vdp = lambda t, x: [x[1], 
                  mu * (1 - x[0] ** 2) * x[1] - x[0]]

IC = [0., 5.]
t0, tf = 0., 16.
nframes = 2 ** 12


soln_harmonic = solve_ivp( f_harmonic, [t0, tf], IC,
                         t_eval=np.linspace(t0, tf, nframes))

soln_vdp = solve_ivp(f_vdp, [t0, tf], IC,
                t_eval=np.linspace(t0,tf,nframes))


# A beautiful!! table in p.167 of 'the periodic solution of 
# van der pol's equation' -- SIAM 2006
# gives high-precision values of T collected from other sources
# for integer values of eps
```

```python
fig, ax = plt.subplots(1, 2, figsize=(7,3.2))



ax[0].plot(soln_harmonic.y[0,:], soln_harmonic.y[1,:])

ax[1].plot(soln_vdp.y[0,:], soln_vdp.y[1,:])

plt.savefig('assets/harmonic_vs_vdp.png', dpi=256, )

plt.show()
```
